```python
def reverse(s):
    if len(s) < 2:
        return s
    else:
      #return s[-1] + reverse(s[:-1]) # Take the last char and put it first
      #return reverse(s[1:]) + s[0] # Take the first char and put it last
      return s[-1] + reverse(s[1:-1]) + s[0] # Take the first and last char and put them in the middle
```
